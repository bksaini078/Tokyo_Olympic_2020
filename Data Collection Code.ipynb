```python
import pandas as pd 
```

```python
df=pd.read_html('https://olympics.com/tokyo-2020/olympic-games/en/results/all-sports/medal-standings.htm')
```

```python
medals_by_country= df[0].rename(columns={'Unnamed: 2':'Gold','Unnamed: 3':'Silver','Unnamed: 4':'Bronze'})
medals_by_country
```

```python
medals_by_country_men= df[1].rename(columns={'Unnamed: 1':'Gold','Unnamed: 2':'Silver','Unnamed: 3':'Bronze'})
medals_by_country_female= df[2].rename(columns={'Unnamed: 1':'Gold','Unnamed: 2':'Silver','Unnamed: 3':'Bronze'})
medals_by_country_mixed= df[3].rename(columns={'Unnamed: 1':'Gold','Unnamed: 2':'Silver','Unnamed: 3':'Bronze'})
medals_by_country_open= df[4].rename(columns={'Unnamed: 1':'Gold','Unnamed: 2':'Silver','Unnamed: 3':'Bronze'})

medals_by_country_open
```

```python
medals_by_country_men['Gender']='Male'
medals_by_country_female['Gender']='Female'
medals_by_country_mixed['Gender']='Mixed'
medals_by_country_open['Gender']='Open'
```

```python
medals_by_country_gender= medals_by_country_men.append(medals_by_country_female).append(medals_by_country_mixed).append(medals_by_country_open)
medals_by_country_gender= medals_by_country_gender.reset_index(drop=True)
```

```python
medals_by_country_gender.to_csv('MedalsByCountryGender.csv',index=False)
```

```python
unique_countries=list(medals_by_country['Team/NOC'].unique())
```

```python
# xdf= pd.read_html('https://olympics.com/tokyo-2020/olympic-games/en/results/all-sports/noc-entries-united-states.htm')
# xdf=xdf[0][:-1]
# xdf
```

```python
discip_country_dict={'United States of America': 'united-states',
                        "People's Republic of China": 'china',
                        "Islamic Republic of Iran":'islamic-rep-of-iran',
                        "Dominican Republic": 'dominican-rep-',
                        "Syrian Arab Republic" :"syrian-arab-rep-",
                        "Republic of Moldova":"rep-of-moldova",
                        "Côte d'Ivoire":"cote-d-ivoire"}

url= 'https://olympics.com/tokyo-2020/olympic-games/en/results/all-sports/noc-entries-'
df=pd.DataFrame()
#getting all country names 
for country in unique_countries:
    if country in discip_country_dict:
        comp_url=url+discip_country_dict[country]+'.htm'
        print(comp_url)
        x=pd.read_html(comp_url)
        x=x[0][:-1]
        x['Team/NOC']=country
        df=df.append(x).reset_index(drop=True)
    else: 
        country= country.lower()
        country= country.replace(' ','-').replace('.','').replace(',','')
        comp_url = url+country+'.htm'
        print(comp_url)
        x=pd.read_html(comp_url)
        x=x[0][:-1]
        x['Team/NOC']=country
        df=df.append(x).reset_index(drop=True)


df
```

```python
discipline_per_country_gender= df
discipline_per_country_gender.to_csv('DisciplineByCountryGender.csv', index=False)
```
